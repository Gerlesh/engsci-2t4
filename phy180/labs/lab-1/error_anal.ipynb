```python
import numpy as np
import matplotlib.pyplot as plt
sum_list = []
j_list = []
for j in np.arange(1,6,0.1):    
    ss = []
    for h in range(1):
        for i in range(int(10**j)):
            t_actual_1 = np.random.random()
            t_actual_2 = np.random.random()
            s.append(t_actual_2-t_actual_1)
        ss.append(sum(s)/len(s))

    sum_list.append(sum(ss)/len(ss))
    j_list.append(j)

# x = np.arange(1, 6, 0.1)
# y = 0.01/np.sqrt(10**x)
# plt.plot(x, y)

plt.plot(j_list, sum_list)
```
