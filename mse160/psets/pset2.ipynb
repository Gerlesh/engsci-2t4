```python
import matplotlib.pyplot as plt
import numpy as np
from intersect import intersection

data = [
[0    , 50.8],
[7.3  , 50.851],
[15.1 , 50.902],
[23.1 , 50.952],
[30.4 , 51.003],
[34.4 , 51.054],
[38.4 , 51.308],
[41.3 , 51.816],
[44.8 , 52.832],
[46.2 , 53.848],
[47.3 , 54.864],
[47.5 , 55.88],
[46.1 , 56.896],
[44.8 , 57.658],
[42.6 , 58.42],
[36.4 , 59.182]]

load = np.zeros(16)
length = np.zeros(16)
stress = np.zeros(16)
strain = np.zeros(16)

length_0 = 50.8
diameter = 12.8
area = np.pi*diameter**2/4
E = 60*1000

for i in range(16):
    load[i] = data[i][0]*1000
    length[i] = data[i][1]

stress = load/area
strain = (length-length_0)/length_0

unload_strain = np.linspace(0.002, 0.009, 40)
unload_stress = unload_strain*E-0.002*E

yield_strain, yield_strength = intersection(strain, stress, unload_strain, unload_stress)

plt.xlabel('Strain')
plt.ylabel('Stress (MPa)')
plt.plot(strain, stress, '-o', color='black', linewidth=1)
plt.plot(unload_strain, unload_stress, color='red', linewidth=1)
plt.plot(yield_strain, yield_strength, "x", color="blue")

plt.savefig('figures/pset2_q5.png', dpi=500)
```

```python
print(f"Yield strength: {yield_strength[0]}")
print(f"Utlimate Tensile Strength: {max(stress)}")
print(f"Fracture Strength: {stress[-1]}")
```
